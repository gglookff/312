```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import matplotlib.animation as animation

class HoopWithMass:
    def __init__(self, R=1.0, m_hoop=0.5, m_mass=0.1, g=9.81):
        self.R = R
        self.m_hoop = m_hoop
        self.m_mass = m_mass
        self.g = g
        self.total_mass = m_hoop + m_mass

    def equations_of_motion(self, t, y):
        x, x_dot, theta, theta_dot = y

        M11 = self.total_mass
        M12 = self.m_mass * self.R * np.cos(theta)
        M22 = self.m_mass * self.R**2

        f1 = self.m_mass * self.R * theta_dot**2 * np.sin(theta)
        f2 = -self.m_mass * self.g * self.R * np.sin(theta)

        det = M11 * M22 - M12**2
        x_ddot = (M22 * f1 - M12 * f2) / det
        theta_ddot = (-M12 * f1 + M11 * f2) / det

        return [x_dot, x_ddot, theta_dot, theta_ddot]

    def simulate(self, initial_conditions, t_span, n_points=1000):
        t_eval = np.linspace(t_span[0], t_span[1], n_points)
        solution = solve_ivp(self.equations_of_motion, t_span, initial_conditions,
                           t_eval=t_eval, method='RK45')

        self.t = solution.t
        self.x = solution.y[0]
        self.x_dot = solution.y[1]
        self.theta = solution.y[2]
        self.theta_dot = solution.y[3]
        self.calculate_energy()

        return solution

    def calculate_energy(self):
        v_hoop_sq = self.x_dot**2
        v_mass_x = self.x_dot + self.R * self.theta_dot * np.cos(self.theta)
        v_mass_y = self.R * self.theta_dot * np.sin(self.theta)
        v_mass_sq = v_mass_x**2 + v_mass_y**2

        kinetic = 0.5 * self.m_hoop * v_hoop_sq + 0.5 * self.m_mass * v_mass_sq
        potential = -self.m_mass * self.g * self.R * np.cos(self.theta)

        self.total_energy = kinetic + potential

    def plot_results(self):
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))

        ax1.plot(self.t, self.x, 'b-', linewidth=2)
        ax1.set_xlabel('Время (с)')
        ax1.set_ylabel('Положение обруча (м)')
        ax1.set_title('Положение центра обруча')
        ax1.grid(True)

        ax2.plot(self.t, np.degrees(self.theta), 'r-', linewidth=2)
        ax2.set_xlabel('Время (с)')
        ax2.set_ylabel('Угол (градусы)')
        ax2.set_title('Угол положения груза')
        ax2.grid(True)

        ax3.plot(self.t, self.x_dot, 'b-', label='Скорость обруча', linewidth=2)
        ax3.plot(self.t, self.theta_dot, 'r-', label='Угловая скорость груза', linewidth=2)
        ax3.set_xlabel('Время (с)')
        ax3.set_ylabel('Скорость')
        ax3.set_title('Скорости')
        ax3.legend()
        ax3.grid(True)

        ax4.plot(self.t, self.total_energy, 'g-', linewidth=2)
        ax4.set_xlabel('Время (с)')
        ax4.set_ylabel('Энергия (Дж)')
        ax4.set_title('Полная энергия системы')
        ax4.grid(True)

        plt.tight_layout()
        plt.show()

# Запуск симуляции
print("Симуляция движения обруча с грузом")
system = HoopWithMass(R=1.0, m_hoop=0.5, m_mass=0.2, g=9.81)
initial_conditions = [0.0, 1.0, 0.1, 0.0]
t_span = [0, 10]

solution = system.simulate(initial_conditions, t_span)
system.plot_results()

print("Максимальное отклонение обруча:", max(abs(system.x)))
print("Максимальный угол груза:", np.degrees(max(abs(system.theta))))
```
